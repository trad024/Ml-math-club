# Single Variable, Single Function

# **Single Variable Function**

In this section, we introduce a simple single-variable function


$$
f(x) = x^2
$$

The first derivative of \( f \) with respect to the variable \( x \) is given by:

$$
\frac{df}{dx} = 2x
$$

To verify this result, we can use **SymPy**. First, we define \( x \) as a mathematical variable using the `symbols` function.


```python
import sympy as sp
x = sp.symbols('x')
f = x**2
f
```

```python
g=sp.exp(x)*sp.cos(x)+sp.sin(x)
g
```

```python
df_dx = sp.diff(f, x )
print("The derivative of f(x) = x² is:")
df_dx
```

```python
dg_dx = sp.diff(g, x)
print("The derivative of g(x)")
dg_dx
```

You know the formula explicitly but don’t really feel like using a pencil to take the derivative  like

```python
x, a, b, c = sp.symbols('x a b c', real=True)
h = sp.exp(-a*sp.sin(x**2)) * sp.sin(b**x) * sp.log(c*sp.sin(x)**2 /x)
h
```

# **Multivariable Single Function**

Multivariable functions require **partial derivatives**. Consider a mapping:

$$
f: \mathbb{R}^2 \to \mathbb{R}, \quad (x, y) \mapsto f(x, y), \quad (x, y) \in \mathbb{R}^2, \; f(x, y) \in \mathbb{R}
$$

For example, let’s take the function:

$$
f(x, y) = 2x y^2 + 3x + 2y
$$

## **Partial Derivatives**

The partial derivatives of \( f \) with respect to \( x \) and \( y \) are:

$$
\frac{\partial f}{\partial x} = 2y^2 + 3
$$

$$
\frac{\partial f}{\partial y} = 4x y + 2
$$

These partial derivatives can be expressed as a **gradient vector**:

$$
\nabla f(x, y) = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right) = \left( 2y^2 + 3, 4x y + 2 \right)
$$

## **Verification Using SymPy**

We can use `sympy` to verify these partial derivatives.


```python
x, y = sp.symbols('x y')
f = 2*x*y**2 + 3*x + 2*y
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
```

```python
(df_dx,df_dy)
```

# **Multivariable Multiple Functions**

In this section, we consider multiple multivariable functions expressed as a **column vector**. Given values for the variables, the output is a vector. For example, consider a two-element column vector where each element is a two-variable function:

$$
\begin{pmatrix}
f_1(x, y) \\
f_2(x, y)
\end{pmatrix}
=
\begin{pmatrix}
3x^2 y + y \\
2x y^2 - 2x
\end{pmatrix}
$$

## **Partial Derivatives and the Jacobian Matrix**

The partial derivatives of both \( f_1 \) and \( f_2 \) can be taken with respect to \( x \) and \( y \). This gives us four partial derivatives, which are expressed as a \( 2 \times 2 \) matrix called the **Jacobian**:

$$
J_f(x, y) =
\begin{pmatrix}
\frac{\partial f_1}{\partial x} & \frac{\partial f_1}{\partial y} \\
\frac{\partial f_2}{\partial x} & \frac{\partial f_2}{\partial y}
\end{pmatrix}
=
\begin{pmatrix}
6x y & 3x^2 + 1 \\
2y^2 - 2 & 4x y
\end{pmatrix}
$$

This **Jacobian matrix** describes the rate of change of the vector-valued function with respect to its input variables.

## **Calculating the Jacobian Using SymPy**

We can calculate the Jacobian using the `sympy` library in Python. First, we define the functions as a column vector using the `Matrix` function. Then, we compute the Jacobian using the `jacobian` method.


```python
x, y = sp.symbols('x y')
f1 = 3*x**2*y + y
f2 = 2*x*y**2 - 2*x
F = sp.Matrix([f1, f2])
X = sp.Matrix([x, y])
J = F.jacobian(X)
print("Jacobian Matrix:")
J
```

# gradient descent function

$$
x_{\text{new}} = x - \alpha \cdot f'(x)
$$

where:  

$$
\begin{aligned}
&x \text{    is the current value,} \\
&\alpha \text{  (learning rate) controls the step size,} \\
&f'(x) \text{ is the gradient (first derivative of } f(x) \text{).}
\end{aligned}
$$


```python
from ipywidgets import interactive
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def df(x):
    return 2 * x

# Gradient descent function
def gradient_descent(starting_point, learning_rate, num_iterations):
    x = starting_point
    history = [x]

    for _ in range(num_iterations):
        gradient = df(x)
        x = x - learning_rate * gradient
        history.append(x)

    return x, history

# Interactive function for gradient descent
def interactive_gradient_descent(iterations, learning_rate):
    starting_point = 10.0
    final_x, history = gradient_descent(starting_point, learning_rate, iterations)

    x_values = np.linspace(-11, 11, 400)
    y_values = f(x_values)

    plt.figure(figsize=(10, 6))
    plt.plot(x_values, y_values, label="f(x) = x²")  # Plot the function
    plt.scatter(history, [f(x) for x in history], color='red', label="Gradient Descent Steps")  # Plot the steps
    plt.title("Gradient Descent on f(x) = x²")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Final value of x: {final_x}")
    print(f"Minimum value of f(x): {f(final_x)}")


interactive_plot = interactive(interactive_gradient_descent, iterations=(1, 50), learning_rate=(0.01, 1.0, 0.01))
interactive_plot
```

The gradient descent for a function of two variables is given by:

$$
\begin{pmatrix}
x_{\text{new}} \\
y_{\text{new}}
\end{pmatrix}
=
\begin{pmatrix}
x \\
y
\end{pmatrix}
-
\alpha
\begin{pmatrix}
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial y}
\end{pmatrix}
$$

where:  

$$
\begin{aligned}
& (x, y) \text{ is the current point,} \\
& \alpha \text{ (learning rate) controls the step size,} \\
& \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \text{ are the partial derivatives of } f(x, y).
\end{aligned}
$$


```python
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    return np.array([2*x, 2*y])

def gradient_descent_multivar(starting_point, learning_rate, num_iterations):
    x, y = starting_point
    history = [(x, y)]

    for _ in range(num_iterations):
        gradient = grad_f(x, y)
        x = x - learning_rate * gradient[0]
        y = y - learning_rate * gradient[1]
        history.append((x, y))

    return (x, y), history

starting_point = (10.0, -8.0)
learning_rate = 0.1
num_iterations = 50

final_point, history = gradient_descent_multivar(starting_point, learning_rate, num_iterations)

history_x = [point[0] for point in history]
history_y = [point[1] for point in history]
history_z = [f(x, y) for x, y in history]

x_values = np.linspace(-11, 11, 100)
y_values = np.linspace(-11, 11, 100)
X, Y = np.meshgrid(x_values, y_values)
Z = f(X, Y)
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap="viridis", alpha=0.6)
ax.plot(history_x, history_y, history_z, 'ro-', label="Gradient Descent Path")
ax.scatter([starting_point[0]], [starting_point[1]], [f(starting_point[0], starting_point[1])], color='blue', label="Start")
ax.scatter([final_point[0]], [final_point[1]], [f(final_point[0], final_point[1])], color='green', label="End (Minimum)")
ax.set_title("Gradient Descent on f(x, y) = x² + y²")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
ax.legend()
plt.show()
print(f"Final point: (x, y) = ({final_point[0]}, {final_point[1]})")
print(f"Minimum value of f(x, y): {f(final_point[0], final_point[1])}")
```
